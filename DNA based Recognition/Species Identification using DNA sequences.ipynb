##Data Processing

```python
datatrain = pd.read_csv('finaltrain_data.csv')
labeltrain = datatrain.iloc[:, 0].values
sequencestrain = datatrain.iloc[:, -1].values
```

```python
datatest = pd.read_csv('finaltest_data.csv')
labeltest = datatest.iloc[:, 0].values
sequencestest = datatest.iloc[:, -1].values
```

###Convert DNA sequences to numerical representations





```python
# Here, a simple encoding is used where A=1, C=2, G=3, T=4
def encode_sequence(sequence):
    encoding = {'A': 1, 'C': 2, 'G': 3, 'T': 4}
    return [encoding[base] for base in sequence]

encoded_sequencestrain = [encode_sequence(seq) for seq in sequencestrain]
encoded_sequencestest = [encode_sequence(seq) for seq in sequencestest]
```

## Convert labels to numerical labels using LabelEncoder


```python
label_encoder = LabelEncoder()
label_encoder.fit(labeltrain)
encoded_labelstrain = label_encoder.transform(labeltrain)
encoded_labelstest = label_encoder.transform(labeltest)
```

```python
np.save('label_encoder.npy', label_encoder.classes_)
```

# Pad sequences to a common length (assuming you're using variable-length sequences)

```python
max_sequence_length = max(len(seq) for seq in encoded_sequencestrain)
X_train_padded = tf.keras.preprocessing.sequence.pad_sequences(encoded_sequencestrain, maxlen=max_sequence_length, padding='post')
X_test_padded = tf.keras.preprocessing.sequence.pad_sequences(encoded_sequencestest, maxlen=max_sequence_length, padding='post')
```

```python
print(max_sequence_length)
```

```python
print(X_train_padded)
```

#Training the Models on dataset

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier =  LogisticRegression(random_state=0)
classifier.fit(X_train_padded,encoded_labelstrain)
```

### Predicting the Test set results

```python
y_pred= classifier.predict(X_test_padded)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print('Accuracy of Logistic Regression is: {0}%'.format(accuracy_score(encoded_labelstest, y_pred)*100))
```
